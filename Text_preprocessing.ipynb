```python
chat_words = {
    "LOL": "Laughing Out Loud",
    "OMG": "Oh My God",
    "BRB": "Be Right Back",
    "TTYL": "Talk To You Later",
    "IMO": "In My Opinion",
    "IDK": "I Don't Know",
    "FYI": "For Your Information",
    "ASAP": "As Soon As Possible"
}
```

```python
def chat_conversion(text):
    
    words = text.split()

    new_text = []

    for word in words:

        if word.upper() in chat_words:

            new_text.append(chat_words[word.upper()])

        else:

            new_text.append(word)

    return " ".join(new_text)
```

```python
text = "LOL I will BRB"

print(chat_conversion(text))
```

```python
def clean_text(text):
    
    # Lowercase
    text = text.lower()

    # Remove HTML
    text = BeautifulSoup(text, "html.parser").get_text()

    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # Expand contractions
    text = contractions.fix(text)

    # Convert emojis
    text = emoji.demojize(text)

    # Remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize using spaCy
    doc = nlp(text)

    # Remove stopwords and lemmatize
    tokens = [
        token.lemma_
        for token in doc
        if not token.is_stop and not token.is_punct
    ]

    return " ".join(tokens)
```

```python
import re
import string
import unicodedata
import contractions
import emoji
import pandas as pd
import spacy
from bs4 import BeautifulSoup
from nltk.corpus import stopwords

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Sample Text
sample = """
<h1>HELLO Everyone!!!</h1>

I can't believe this Movie is AMAZING 😍😍

Visit https://google.com

Email: abc123@gmail.com

LOL 😂😂

#AI @OpenAI
"""

# Dictionary to store each preprocessing step
steps = {}

# Original
steps["Original"] = sample

# 1. Lowercase
text = sample.lower()
steps["1. Lowercase"] = text

# 2. Unicode Normalization
text = unicodedata.normalize("NFKD", text).encode("ascii", "ignore").decode("utf-8")
steps["2. Unicode Normalization"] = text

# 3. Remove HTML Tags
text = BeautifulSoup(text, "html.parser").get_text()
steps["3. Remove HTML"] = text

# 4. Remove URLs
text = re.sub(r'https?://\S+|www\.\S+', '', text)
steps["4. Remove URLs"] = text

# 5. Remove Email IDs
text = re.sub(r'\S+@\S+', '', text)
steps["5. Remove Emails"] = text

# 6. Expand Contractions
text = contractions.fix(text)
steps["6. Expand Contractions"] = text

# 7. Remove Mentions
text = re.sub(r'@\w+', '', text)
steps["7. Remove Mentions"] = text

# 8. Remove Hashtags (# only)
text = re.sub(r'#', '', text)
steps["8. Remove Hashtags"] = text

# 9. Convert Emojis to Text
text = emoji.demojize(text)
steps["9. Emoji to Text"] = text

# 10. Remove Punctuation
text = text.translate(str.maketrans("", "", string.punctuation))
steps["10. Remove Punctuation"] = text

# 11. Remove Numbers
text = re.sub(r'\d+', '', text)
steps["11. Remove Numbers"] = text

# 12. Remove Extra Spaces
text = re.sub(r'\s+', ' ', text).strip()
steps["12. Remove Extra Spaces"] = text

# 13. Tokenization (spaCy)
doc = nlp(text)
tokens = [token.text for token in doc]
steps["13. Tokenization"] = str(tokens)

# 14. Stopword Removal
stop_words = set(stopwords.words("english"))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
steps["14. Stopword Removal"] = str(filtered_tokens)

# 15. Lemmatization
doc = nlp(" ".join(filtered_tokens))
lemmas = [token.lemma_ for token in doc]
steps["15. Lemmatization"] = str(lemmas)

# Final Clean Text
steps["16. Final Output"] = " ".join(lemmas)

# Display Result
df = pd.DataFrame(steps.items(), columns=["Preprocessing Step", "Output"])

df
```
